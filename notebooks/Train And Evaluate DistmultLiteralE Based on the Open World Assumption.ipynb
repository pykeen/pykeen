```python
import logging
import os
import sys
import time
import poem
from poem.constants import KG_EMBEDDING_MODEL_NAME, DISTMULT_LITERAL_NAME_OWA, TRAINING_SET_PATH, PATH_TO_NUMERIC_LITERALS,KG_ASSUMPTION,OWA,EMBEDDING_DIM,MARGIN_LOSS,OPTMIZER_NAME,SGD_OPTIMIZER_NAME, LEARNING_RATE, NUM_EPOCHS, BATCH_SIZE, EVALUATOR,RANK_BASED_EVALUATOR,SEED, NUM_ENTITIES, NUM_RELATIONS
from poem.pipeline import Helper
from poem.preprocessing.triples_preprocessing_utils.basic_triple_utils import create_entity_and_relation_mappings, load_triples
from poem.instance_creation_factories.triples_numeric_literals_factory import TriplesNumericLiteralsFactory
from poem.model_config import ModelConfig
from poem.kge_models.kge_models_using_numerical_literals.distmult_literal_e_owa import DistMultLiteral
from poem.evaluation.ranked_based_evaluator import RankBasedEvaluator
import numpy as np
```

Define datapaths:

```python
triples_path = '../tests/resources/test.txt'
literals_path = '../tests/resources/numerical_literals.txt'
```

Define config:

```python
config = {
        KG_ASSUMPTION: OWA,
        EMBEDDING_DIM: 100,
        MARGIN_LOSS: 1,
        OPTMIZER_NAME: SGD_OPTIMIZER_NAME,
        LEARNING_RATE: 0.01,
        NUM_EPOCHS: 1,
        BATCH_SIZE: 128,
    }
```

Create instances:

```python
triples = load_triples(path=triples_path)
numeric_triples = load_triples(path=literals_path)
entity_to_id, relation_to_id = create_entity_and_relation_mappings(triples=triples)

factory = TriplesNumericLiteralsFactory(
        entity_to_id=entity_to_id,
        relation_to_id=relation_to_id,
        numeric_triples=numeric_triples
    )
instances = factory.create_owa_instances(triples=triples)
```

Define number of entities and relations

```python
config[NUM_ENTITIES] = len(instances.entity_to_id)
config[NUM_RELATIONS] = len(instances.relation_to_id)
```

Create model config:

```python
model_config = ModelConfig(config=config, multimodal_data=instances.multimodal_data)
```

Define KGE model:

```python
kge_model = DistMultLiteral(model_config=model_config)
```

Configure training loop:

```python
all_entities = np.array(list(entity_to_id.values()))
```

```python
training_loop = Helper.get_training_loop(config=config,
                                         kge_model=kge_model,
                                         all_entities=all_entities)
```

Train KGE model:

```python
fitted_kge_model, losses = training_loop.train(training_instances=instances)
```

Get evaluator: 

```python
evaluator = RankBasedEvaluator(
    kge_model=fitted_kge_model,
    entity_to_id=entity_to_id,
    relation_to_id=relation_to_id,
    training_triples=instances.instances,
    filter_neg_triples=False)
```

Evaluate KGE model:

```python
metric_results = evaluator.evaluate(test_triples=instances.instances[0:100])
```
